```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv('imdb-top-1000.csv')
movies.sample()
```

```python
genres = movies.groupby('Genre')
genres
```

```python
# Applying Builtin aggregation functions on groupby object
genres.sum()
```

```python
# find the top 3 genres by total earnings
movies.groupby('Genre').sum()['Gross'].sort_values(ascending=False).head(3)
```

```python
movies.groupby('Genre')['Gross'].sum().sort_values(ascending=False).head(3)
# this one is more efficient, because sum is applying only in one column called Gross  
```

```python
# find the genres which have highest avg IMDB ratings
genres['IMDB_Rating'].mean().sort_values(ascending= False).head(1)
```

```python
# director with the most popularity
movies.groupby('Director')['No_of_Votes'].sum().sort_values(ascending=False).head()
```

```python
# Find the highest rated movie of each genres
movies.groupby('Genre')['IMDB_Rating'].max().sort_values(ascending=False).head()
```

```python
# find no of movies done by each actors
# movies['Star1'].value_counts()
movies.groupby('Star1')['Series_Title'].count().sort_values(ascending=False)
```

```python
# GroupBy Attributes and Methods
# find total number of groups -> len
# find items in each group -> size
# first()/last() -> nth item
# get_group -> vs filtering
# groups
# describe
# sample
# nunique
```
