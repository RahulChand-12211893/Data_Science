```python
#calling diffrenet function
lst = [1,2,3,4,5,6]
def squ_lst(lst):
    return [i**2 for i in lst]
def cube_lst(lst):
    return [i**3 for i in lst]
def final(lst):
    lst1 = squ_lst(lst)
    lst2 = cube_lst(lst)
    return [lst1[i] + lst2[i] for i in range(len(lst1))]
print(squ_lst(lst))
print(cube_lst(lst))
print(final(lst))
```
